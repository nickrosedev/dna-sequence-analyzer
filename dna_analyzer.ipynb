```python
from Bio import SeqIO
import matplotlib.pyplot as plt

# Load the FASTA file
fasta_file = "ncbi_dataset/data/gene.fna"
record = SeqIO.read(fasta_file, "fasta")
sequence = record.seq

print(f"Loaded sequence: {record.id}")
print(f"Length: {len(sequence)} nucleotides")
```

```python
# Count each nucleotide
count_A = sequence.count('A')
count_T = sequence.count('T')
count_C = sequence.count('C')
count_G = sequence.count('G')

print(f"A: {count_A}")
print(f"T: {count_T}")
print(f"C: {count_C}")
print(f"G: {count_G}")
print(f"Total: {count_A + count_T + count_C + count_G}")
```

```python
# GC content is the percentage of G and C bases
gc_content = (count_G + count_C) / len(sequence) * 100
print(f"GC Content: {gc_content:.2f}%")
```

```python
# Find all positions where ATG appears
motif = "ATG"
positions = []
for i in range(len(sequence) - 2):
    if sequence[i:i+3] == motif:
        positions.append(i)

print(f"Found {len(positions)} occurrences of '{motif}'")
print(f"First 10 positions: {positions[:10]}")
```

```python
# Bar chart of nucleotide counts
nucleotides = ['A', 'T', 'C', 'G']
counts = [count_A, count_T, count_C, count_G]

plt.figure(figsize=(8, 5))
plt.bar(nucleotides, counts, color=['green', 'red', 'blue', 'orange'])
plt.xlabel('Nucleotide')
plt.ylabel('Count')
plt.title('Nucleotide Frequency Distribution')
plt.show()
```
